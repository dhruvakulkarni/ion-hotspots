```python
import numpy as np

from geometry import Geometry
from material import Material
from srim_output import SRIMOutput
from ion_source import GaussianIonBeam

from simulation import run_simulation  
from plotting import plot_voxels, plot_field_slice, plot_field_hist
from plotting import plot_field_3d_heatmap
from plotting import plot_field_3d_interactive, plot_field_3d_cells_plotly

# ============================================================
# MAIN
# ============================================================

def main():

    # -------------------------
    # Geometry (thin film system)
    # -------------------------
    geometry = Geometry(
        lx=1e-6,
        ly=1e-6,
        lz=200e-9,
        nx=30,
        ny=30,
        nz=10
    )

    # -------------------------
    # Material
    # -------------------------
    material = Material(
        name="NbN",
        density=8400,
        stopping_power_scale=1.0,
        cylinder_model_radius=10e-9
    )

    # -------------------------
    # SRIM model (He ion)
    # -------------------------
    srim = SRIMOutput(
        ion_name="He",
        ion_energy_eV=5.3e6,
        ion_mass_amu=4.0,
        dEdx_electronic_eVperA=3.510e1,
        dEdx_nuclear_eVperA=2.727e-2,
        range_projected=5e-6,   
        straggle_long=500e-9,     
        straggle_trans=450e-9
    )

    # -------------------------
    # Ion beam (Gaussian irradiation)
    # -------------------------
    beam = GaussianIonBeam(
        current_A=1e-12,
        energy_eV=5.3e6,
        mass_amu=4.0,
        sigma_x=100e-9,
        sigma_y=100e-9,
        #angular_spread=0.01
    )

    # -------------------------
    # Run full simulation (ENGINE CALL)
    # -------------------------
    t_end = 1e-6

    energy,temperature = run_simulation(
        ion_source=beam,
        geometry=geometry,
        material=material,
        srim=srim,
        t_end=t_end,
        max_events=3
    )

    # -------------------------
    # Analysis
    # -------------------------
    #mask = field > 1e-3 * field.max()
    mask = energy > 0

    print("\n[RESULTS]")
    print("Field shape:", energy.shape)
    print("Max energy density:", energy.max())
    print("Affected voxels:", np.sum(mask))

    # -------------------------
    # Visualization
    # -------------------------
    plot_voxels(geometry, mask)
    plot_field_hist(energy)
    plot_field_hist(temperature)

    plot_field_slice(
        geometry,
        energy,
        axis="z",
        title="SRIM Energy Deposition (z-slice)"
    )
    
    plot_field_3d_heatmap(
    geometry,
    energy,
    threshold=0.0
    )
    
    plot_field_3d_interactive(
        geometry,
        energy,
        threshold=0.0,
        label='energy'
    )
    plot_field_3d_interactive(
        geometry,
        temperature,
        threshold=0.0,
        label='temp'
    )
    
    plot_field_3d_cells_plotly(
        geometry,
        energy,
        threshold=1e-3 * field.max()
    )

    
    
```

```python
main()
```
