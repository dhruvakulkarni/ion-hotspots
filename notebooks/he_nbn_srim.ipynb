```python
from srim_output import SRIMOutput


# ============================================================
# SRIM OUTPUT:
# 5.3 MeV He in NbN
# ============================================================

he_in_nbn_srim = SRIMOutput(

    # ion properties
    ion_name="He",
    ion_energy_eV=5.3e6,
    ion_mass_amu=4.003,

    # stopping powers
    #
    # SRIM units:
    #   keV / micron
    #
    # Convert:
    #   1 keV/micron =
    #   1e3 * 1.602e-19 / 1e-6
    #   = 1.602e-10 J/m
    #

    dEdx_electronic=(
        3.510e2 * 1.602e-10
    ),

    dEdx_nuclear=(
        2.727e-1 * 1.602e-10
    ),

    # projected range
    #
    # 10.18 micron → meters
    #
    range=10.18e-6,

    # longitudinal straggling
    #
    # 4503 A → meters
    #
    straggle_long=4503e-10,

    # lateral straggling
    #
    # 5510 A → meters
    #
    straggle_trans=5510e-10,
)
```
